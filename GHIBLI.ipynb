```python
# Define paths (You can change input_path and output_path as needed)
input_path = "/content/drive/MyDrive/PROJECTS/GHIBLI ART GENERATION/sample5.jpeg"  # Your image file path
output_path = "/content/drive/MyDrive/PROJECTS/GHIBLI ART GENERATION/output5.jpeg"

# Load the input image (if you're processing a new image)
init_image = Image.open(input_path).convert("RGB")
init_image = init_image.resize((512, 512))  # Resize for model compatibility

# Define your prompt
prompt = "A young woman standing on a countryside path surrounded by green fields and blue skies, wearing a flowing dress, Studio Ghibli style, soft pastel colors, dreamy and magical atmosphere, delicate watercolor effect, highly detailed environment"

pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "hakurei/waifu-diffusion",
    torch_dtype=torch.float16,
).to("cuda")
# Generate the Ghibli Art
image = pipe(
    prompt=prompt,
    image=init_image,
    strength=0.6,        # Strength of transformation
    guidance_scale=7.5   # Influence of the prompt
).images[0]

# Save the output image
image.save(output_path)

# Display the output
image.show()

print("\n✨ Ghibli Art generated and saved successfully! ✨")
```
